```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
users=pd.read_csv('BX-Users.csv',sep=';',encoding='latin-1',error_bad_lines=False)
books=pd.read_csv('BX-Books.csv',sep=';',encoding='latin-1',error_bad_lines=False)
ratings=pd.read_csv('BX-Book-Ratings.csv',sep=';',encoding='latin-1',error_bad_lines=False)
```

```python
print(users.shape)
print(books.shape)
print(ratings.shape)
```

```python
# Pivot Table for better visualization of data
ratings.head(10)
```

```python
ratings.rename(columns={'User-ID':'user-id','Book-Rating':'score'},inplace=True)
ratings.head(10)
```

```python
x=ratings['user-id'].value_counts()>200
padhaku=x[x].index.tolist()
padhaku
```

```python
ratings=ratings[ratings['user-id'].isin(padhaku)]
print(ratings.shape)
print(len(padhaku))
```

```python
ratings['ISBN'].nunique()
```

```python
books=books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
```

```python
books.head(10)
```

```python
books_with_ratings=ratings.merge(books,on='ISBN')
```

```python
books_with_ratings.head(56)
```

```python
a=books_with_ratings['ISBN'].value_counts()>50
famous=a[a].index.tolist()
len(famous)
```

```python
books_with_ratings=books_with_ratings[books_with_ratings['ISBN'].isin(famous)]
books_with_ratings.shape
```

```python
books_with_ratings.drop_duplicates(subset=['user-id','Book-Title'],inplace=True)
```

```python
books_with_ratings.shape
```

```python
pt=books_with_ratings.pivot_table(index='Book-Title',columns='user-id',values='score')
pt.shape
```

```python
pt.fillna(-1,inplace=True)
pt
```

```python
from scipy.sparse import csr_matrix
```

```python
matrix=csr_matrix(pt)
```

```python
from sklearn.neighbors import NearestNeighbors
NN=NearestNeighbors(algorithm='brute')
```

```python
NN.fit(matrix)
```

```python
pt.index[0]
```

```python
pt.iloc[0,:].values.reshape(1,883)
```

```python
d,s=NN.kneighbors(pt.iloc[153,:].values.reshape(1,883),n_neighbors=5)
```
